## 1. Loading the data

When calling `fetch_20newsgroups` we can either get all documents in the set or have subsets split for training and testing. To explore the set we will initially load it as a whole. 

We also remove the assosiated document metadata.

## 4. Thematic Structure 

Unfortunately the heyday of **`USENET`**, the network hosting these newsgroups is long gone and so are many of its newsgroups. The *20 Newsgroups* dataset was [allegedly](http://qwone.com/~jason/20Newsgroups/) collected by Ken Lang in the 1990's for his paper on *Newsweeder: Learning to filter netnews*. According to a source cited by [www.usenet.com](https://www.usenet.com/what-is-a-newsgroup-and-how-do-i-use-them/) currently, there are only 20 000/110 000 active newsgroups. 

[This article](https://www.ou.edu/research/electron/internet/use-faq.htm), titled *"What is Usenet?", 1995* and posted on the `news.announce.newusers, news.admin.misc` and `news.answers` newsgroups and [this instructional manual](https://www.whitman.edu/mathematics/eegtti/eeg_70.html), last updated in 1991, both tell us that there was a guide available on the network regarding the launching of new newsgroups and that newsgroups were only created when necessary and approved by a popular vote. 

Today, [www.usenet.com](https://www.usenet.com/usenet-faq/)'s FAQ page offers some insight how the topic hierarchies are ogranised at the highest level:

>*Within the Usenet world, all newsgroups are organized into nine different categories:*
>```
>comp.* (computer related discussions)
>humanities.* (philosophy, literature and arts discussions)
>misc.* (miscellaneous discussions)
>news.* (news discussions)
>rec.* (recreational and entertainment discussions)
>sci.* (science discussions)
>soc.* (social discussions)
>talk.* (controversial topic discussions)
>alt.* (alternative discussions)
>```
>

A thing to [note](https://www.usenet.com/what-are-news-groups-hierarchy-the-big-8/) here is the difference between `alt.` and `misc.` categories, where `misc.` covers niche topics and `alt.` allows you to build a wide-scope topic with no restrictions or moderation of the newsgroup.

We can see that the *20 newsgroups* dataset includes messages from all topics, apart from `news.` and `humanities.` We can get a deeper understanding of the hierarchical structure of the corpus by plotting several tree diagrams. 

### 4.1. Topic Tree Diagrams

We define a few functions to generate tree diagram visualisations of the corpus' topic structure, so we can further analyse.

````python
# Build the tree

def build_tree(categories):
    # Initialize tree using lambda function to create a nested defaultdict
    tree = lambda: defaultdict(tree)
    root = tree()
    for category in categories:
        parts = category.split('.')
        current_level = root
        for part in parts:
            current_level = current_level[part]
    return root
````

````python
# Adding edges to a graph

def add_edges(graph, node, subtree):
    for child in subtree:
        child_node = f"{node}.{child}" if node else child
        graph.add_edge(node, child_node)
        add_edges(graph, child_node, subtree[child])
````

````python
# Drawing a tree graph

def plot_tree(subtree, root_label, ax, k=0.1, iterations=50):
    G = nx.DiGraph()
    add_edges(G, root_label, subtree)
    
    pos = nx.spring_layout(G, seed=42, k=k, iterations=iterations)
    
    # Manually adjust the position of the 'talk.religion.christian' node
    if 'talk.religion.christian' in pos:
        pos['talk.religion.christian'][0] += 1.5  # quick-fix to improve readability
        pos['talk.religion.christian'][1] += 0.2  

    
    # Adjust labels to only show the node name
    labels = {node: node.split('.')[-1] for node in G.nodes}

    # Draw the graph with the center node colored pink
    nx.draw(G, pos, labels=labels, 
            node_size=2000, node_color='skyblue', font_size=12, font_weight='bold', edge_color='gray', ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=[root_label], node_color='pink', node_size=3000, ax=ax)
````

And then produce the visualisation:

````python
categories = np.array(newsgroups_df["target_name"].unique())

# Build the tree
root_raw = build_tree(categories)

# Create a 2x2 subplot to draw the first 4 trees
fig, axs = plt.subplots(2, 2, figsize=(16, 16))
fig.suptitle("\nTopic Tree Diagrams", fontsize=36, fontweight='bold')  # Adjust the y parameter for padding


# Plot each root node's subtree in the 2x2 grid
root_nodes = list(root_raw.keys())[:4]
for ax, root_node in zip(axs.flatten(), root_nodes):
    plot_tree(root_raw[root_node], root_node, ax, k=0.35, iterations=50)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

<center><sub>Fig.1 The tree diagrams of the four super-topics.</sub></center>

&nbsp;

### 4.2. Structure Analysis

#### 4.2.1. Niche topics
Starting with the topics not shown in Fig.1: `alt.atheism`, `misc.forsale` and `soc.religion.christian`, we can see that one is talking about *christianity*, one about *atheism* and the third one is a "for-sale" advertising ground. There is, however, another topic, i.e. `talk.religion.misc` which also covers the subject of religion. 

`talk.religion.misc` is the smallest topic with 628 documents and is also the vaguest. It is safe to assume that any topic modelling performed will inevitably mix these religion-related topics in some way. It is possible that a model might not link religion to politics as a neighbouring topic.

Additionally the `misc.forsale` topic might introduce noise to the model. For example, if there is a big number of adverts selling sports goods, the model might falsely classify these topics as `rec.sport` topics, however this still remains somewhat semantically true ([rec.auto](https://www.ibiblio.org/rdu/usenet-arc/sub00043/msg00000.html) forbids the postings of for-sale ads). 


#### 4.2.2 Super Topics (Themes)
We can find the `rec.` topics in the diagram at the top left. The sub-topics of `rec.` are `rec.auto` (cars), `rec.motorcycles` and `rec.sport`, which is further divided into `rec.sport.baseball` and `rec.sport.hockey`. 

Semantically, while _baseball_ and _hockey_ vocabulary can share generic sport terms such as: *player, captain, team, game, fans, coach, home, away;* they can also differ a lot in areas such as the teams' unique names (probably discussed a lot in an online forum), sport-specific terminology such as the playground, equipment, leagues, etc. 

_Cars_ and _motorcycles_, however, intuitively share a lot more of the same lexicon, so the two topics might mix, which is something to keep an eye on when observing the output of the LDA model.

The `comp.` diagram seems complicated at first glance, but can be simplified into 4 subtopics:
1. one about Windows X (a version launched in the 80s),
2. one about miscellaneous discussions about Windows OS,
3. one about computer graphics and
4. a final one about hardware, consisting of a further split between Mac and IBM PC's. 

Even though in the tree diagram the Windows nodes are separated, one being a child `os.`, realistically the thematic split is between the narrow topic of *Windows X* and the broader `misc.` topic. 

The `talk.` theme is really mostly about politics with the small `religion.mis` that we already covered. As we unfortunately know, [*guns* and *middle east* have some common ground](https://rkba.org/research/rkba.faq), although `politics.guns` will most probably mainly revolve around gun control in the US. 

As with religion, `politics.misc` might spill over into one of the other political topics. Generally, the over-arching theme `talk.` is also ambiguous, when the documnents in question extracted from discussuion forums. We can suspect, that, the purpose for this newsgroup naming branch has to do with the frequency, lenght and lexicon (shorter, more basic, but frequent discussions). 

Finally, the science group is the simplest, most clearly defined, with only possibility of mixing between `comp.` documents and `sci.electronics`. 

We can revise the topical hierarchy like so: 

````python
revised_categories = np.array(['rec.sport.hockey', 'comp.hardware.ibm-pc',
                       'talk.politics.mideast', 'comp.hardware.mac',
                       'sci.electronics', 'talk.religion.misc', 'sci.crypt', 'sci.med',
                       'talk.religion.atheism', 'rec.motorcycles', 'rec.cars', 'comp.windows.x',
                       'comp.graphics', 'sci.space', 'talk.politics.guns', 'misc.forsale',
                       'rec.sport.baseball', 'talk.politics.misc',
                       'comp.windows.misc', 'talk.religion.christian'])
````

And then produce the new tree diagrams:

````python
# Build the tree
root_revised = build_tree(revised_categories)

# Create a 2x2 subplot to draw the first 4 trees
fig, axs = plt.subplots(2, 2, figsize=(16, 16))
fig.suptitle("\nRevised Topic Tree Diagrams", fontsize=36, fontweight='bold')  # Adjust the y parameter for padding


# Plot each root node's subtree in the 2x2 grid
root_nodes = list(root_revised.keys())[:4]
for ax, root_node in zip(axs.flatten(), root_nodes):
    plot_tree(root_revised[root_node], root_node, ax, k=0.35, iterations=100)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

<center><sub>Fig.2 The revised tree diagrams of the four super-topics.</sub></center>

## 5. Document Structure 

We can also investigate how the documents are structured based on their topics and themes. We will first have a look at the documents' length. This is important as LDA is a probability-based model, so longer documents would suggest better overall topic interpretability. 

We can change the target names to our revised categories and then add the document length and the high-level topic, i.e. the theme, defined by the string before the first `'.'` separator. We can  also calculate the average document length of every topic.

````python
# Replace the target_name column using mapping
category_mapping = dict(zip(categories, revised_categories))
newsgroups_df['target_name'] = newsgroups_df['target_name'].replace(category_mapping)

# Compute the length of each document and add it to the DataFrame
newsgroups_df['doc_length'] = newsgroups_df['document'].apply(len)

# Extract the string before the first '.' and create a new column 'target_theme'
newsgroups_df['target_theme'] = newsgroups_df['target_name'].apply(lambda x: x.split('.')[0])

# Group by target_name and calculate the average document length per category
average_lengths = newsgroups_df.groupby('target_name')['doc_length'].mean().round(2)
mean_doc_lengths = average_lengths.groupby(average_lengths.index.str.split('.').str[0]).mean().round(2)

display(Markdown("### 5.1. Newsgroup documents containing mean n words:")) 
print(average_lengths)

display(Markdown("### 5.2. Theme documents containing mean n words:")) 
print(mean_doc_lengths)
````

We can plot a bar chart to get a better idea of the mean word count distribution among topics.

````python
# Function to plot a bar chart
def plot_bar_chart(data):
    # Extract theme names (the string before the first ".")
    themes = data.index.str.split('.').str[0]
    
    # Create a unique color for each theme
    unique_themes = themes.unique()
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_themes)))
    color_map = dict(zip(unique_themes, colors))
    
    # Map the colors to the themes
    bar_colors = themes.map(color_map)
    
    # Group by theme and calculate the mean document length per theme
    mean_doc_lengths = data.groupby(themes).mean().round(2)
    
    # Create the bar chart
    plt.figure(figsize=(14, 8))
    bars = plt.barh(data.index, data.values, color=bar_colors)
    plt.xlabel('Mean Document Length', fontsize=16)
    plt.ylabel('Newsgroup', fontsize=16)
    plt.tight_layout()
    
    # Annotate each bar with the document length
    for bar, value in zip(bars, data.values):
        plt.text(value, bar.get_y() + bar.get_height() / 2, f'{value:.0f}', va='center', ha='right')
    
    # Create a legend
    handles = [plt.Rectangle((0,0),1,1, color=color_map[theme]) for theme in unique_themes]
    legend_labels = [f'{theme} ({mean_doc_lengths[theme]:.0f})' for theme in unique_themes]
    handles.reverse()
    legend_labels.reverse()
    plt.legend(handles, legend_labels, title='Theme (Mean Doc Length)')
    
    # Show the plot
    plt.show()
````

````python
# Use the function for the original data
plot_bar_chart(average_lengths)
````

<center><sub>Fig.3 Mean Document Length per Topic.</sub></center>

&nbsp;

Contrary to our initial presumption, the `talk.` theme has the lead in document length by a big margin. Talks about politics in the middle east, a very hot issue in the US in the 90's, have an average length of a 5-page A4 document. Following are discussions about computers and science with about an even split and finally the for-sale forum and recreational topics. 

We can plot the themes in a box plot to get a bit more insight into the thematic document length distributions. 

````python
def plot_thematic_boxplots(data):
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(12, 8))

    # Extract unique themes
    themes = data['target_theme'].unique()

    # Create box plots for each theme
    data_to_plot = [data[data['target_theme'] == theme]['doc_length'] for theme in themes]
    ax.boxplot(data_to_plot, patch_artist=True)

    # Set y label
    ax.set_ylabel('Document Length', fontsize=16)

    # Set x-axis tick labels
    ax.set_xticks(range(1, len(themes) + 1))
    ax.set_xticklabels(themes, rotation=45, ha='right')

    # Display the plot
    plt.tight_layout()
    plt.show()
````

````python
plot_thematic_boxplots(newsgroups_df)
````

<center><sub>Fig.4 Mean Document Length per Theme.</sub></center>

&nbsp;

### 5.3 Removing Outliers

We can see that document word count distributions across all themes are heavily skewed by long publications (could be actual books, articles, essays and other forms of literature; in the extreme case of one `comp.` document the length is in the 160 000s, an equivalent of hundreds of pages). We can write a function to remove the outliers from the dataframe and then plot the data again to see how it compares. 

````python
# Remove outliers and return the IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    filtered_df = df[(df[column] >= (Q1 - 1.5 * IQR)) & (df[column] <= (Q3 + 1.5 * IQR))]
    return filtered_df
````

````python
newsgroups_df_fltr = remove_outliers(newsgroups_df, "doc_length")
display(Markdown(f"After filtering the data we now have {len(newsgroups_df_fltr)}/{len(newsgroups_df)} documents left."))
````

````python
# Use the function for the filtered data
average_lengths_fltr = newsgroups_df_fltr.groupby('target_name')['doc_length'].mean().round(2)
plot_bar_chart(average_lengths_fltr)
````

<center><sub>Fig.4 Mean Document Length per Topic (outliers removed).</sub></center>

&nbsp;

We can see that the variance between themes has evened out, however the average length for `sci.` documents has remained higher than the `comp.` theme, whereas before removing outliers the difference was relatively smaller.

To compare both cases we can plot both distributions in the same chart.

````python
#Decrease the brightness of a given color
def mute_color(color, factor=0.8):
    rgb = mcolors.to_rgb(color)
    hsv = mcolors.rgb_to_hsv(rgb)
    hsv[2] *= factor  
    muted_rgb = mcolors.hsv_to_rgb(hsv)
    return muted_rgb

def plot_overlaid_bar_chart(data_raw, data_fltr):
    # Extract theme names (the string before the first ".")
    themes = data_raw.index.str.split('.').str[0]
    
    # Create a unique color for each theme
    unique_themes = themes.unique()
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_themes)))
    color_map = dict(zip(unique_themes, colors))
    
    # Create muted colors for data_raw
    muted_color_map = {theme: mute_color(color_map[theme]) for theme in unique_themes}
    
    # Map the colors to the themes for the bars
    bar_colors_raw = themes.map(muted_color_map)
    bar_colors_fltr = themes.map(color_map)
    
    # Group by theme and calculate the mean document length per theme
    mean_doc_lengths_raw = data_raw.groupby(themes).mean().round(2)
    mean_doc_lengths_fltr = data_fltr.groupby(themes).mean().round(2)
    
    # Number of bars
    n_bars = len(data_raw)
    
    # Create a figure and a set of subplots
    fig, ax = plt.subplots(figsize=(14, 10))
    
    # Bar width
    width = 0.8
    
    # Positions of the bars on the y-axis
    pos_bars = np.arange(n_bars)
    
    # Plot the bars
    bars_raw = ax.barh(pos_bars, data_raw.values, width, color=bar_colors_raw)
    bars_fltr = ax.barh(pos_bars, data_fltr.values, width / 2, color=bar_colors_fltr)
    
    # Add labels and legend
    ax.set_xlabel('Document Length', fontsize=16)
    ax.set_ylabel('Newsgroup', fontsize=16)
    ax.set_yticks(pos_bars)
    ax.set_yticklabels(data_raw.index)
    
    # Annotate each bar with the mean document length
    # Raw
    for bar, value_raw, value_fltr in zip(bars_raw, data_raw.values, data_fltr.values):
        ax.text(value_raw + 75, bar.get_y() + bar.get_height() / 2, f'{value_raw:.0f}', va='center', ha='right', fontsize=10)
    
    # Filtered
    for bar, value_raw, value_fltr in zip(bars_fltr, data_raw.values, data_fltr.values):
        ax.text(value_fltr, bar.get_y() + bar.get_height() / 2, f'{value_fltr:.0f}', va='center', ha='right', fontsize=10)
    
    # Create a legend with mean document lengths per theme
    handles = [plt.Rectangle((0,0),1,1, color=color_map[theme]) for theme in unique_themes]
    legend_labels = [f'{theme} (Raw: {mean_doc_lengths_raw[theme]:.0f}, Fltr: {mean_doc_lengths_fltr[theme]:.0f})' for theme in unique_themes]
    
    # Place the legend inside the plot area where there is space
    handles.reverse()
    legend_labels.reverse()
    ax.legend(handles, legend_labels, title='Theme (Mean Doc Length)', loc='best')
    
    # Show the plot
    plt.tight_layout()
    plt.show()
````

````python
plot_overlaid_bar_chart(average_lengths, average_lengths_fltr)
````

<center><sub>Fig.5 Mean Document Length per Topic (filtered data overlaid).</sub></center>

&nbsp;

We can note the few topics where the drop in average document lenght, after removing outliers: `sci.electornics`, most of the  `rec.` theme, the `comp.harware` sub-theme. 

Finally we can plot the theme's again in box plots after removing the outliers.

````python
plot_thematic_boxplots(newsgroups_df_fltr)
````

<center><sub>Fig.6 Mean Document Length per Theme (outliers removed).</sub></center>

&nbsp;

We can see that **Q1** across all themes sits at about **200** words, i.e. 1-2 paragraphs of written text. **Q3** seems to lie around the **700** word mark for most themes, apart from `talk.` and `sci.`, where the **IQR** is slightly larger, indicating that science, religion and politics are prone to lengthier posts. We can see that there are still a number of articles that sit beyond the 1.5 x **Q3** mark, which for all but the aforementioned themes is at about **1500** words.
