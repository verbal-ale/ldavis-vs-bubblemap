### 1.1. Initial  Vocabulary Formation


````python
tfidf_vectorizer  = TfidfVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)
dtm_tfidf  = tfidf_vectorizer.fit_transform(newsgroups_df['document'].fillna('')) # replaces na values with empty strings
print(f"Document-term matrix: {dtm_tfidf.shape[0]} documents x {dtm_tfidf.shape[1]} terms.")
````

In the previous cell we produced a matrix where one dimension contains the documents themselves and the other terms from the vocabulary we have generated using _SciKitLearn's_ [TfidfVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html#tfidfvectorizer):

#### TfidfVectorizer

    Convert a collection of raw documents to a matrix of TF-IDF features.

    Equivalent to CountVectorizer followed by TfidfTransformer.

The application of the ```TfidfTransformer``` is important as it reduces the weight frequently used generic terms carry when forming the topics. We can now fit the model to this matrix.

| |
|:--|
|**strip_accents : _{‘ascii’, ‘unicode’} or callable, default=None_**|
|```Remove accents and perform other character normalization during the preprocessing step. ‘unicode’ is a slightly slower method that works on any characters.```|    
|**stop_words : _{‘english’}, list, default=None_**|
|```‘english’, a built-in stop word list for English is used. There are several known issues with ‘english’ and you should consider an alternative (see Using stop words).```|
|**lowercase  : _bool, default=True_**|
|```Convert all characters to lowercase before tokenizing.```|
|**token_pattern : _str or None, default=r”(?u)\b\w\w+\b”_**|
|```Regular expression denoting what constitutes a “token”, only used if analyzer == 'word'. The default regexp select tokens of 2 or more alphanumeric characters (punctuation is completely ignored and always treated as a token separator). If there is a capturing group in token_pattern then the captured group content, not the entire match, becomes the token. At most one capturing group is permitted.```|
|**max_df : _float in range [0.0, 1.0] or int, default=1.0_**|
|```When building the vocabulary ignore terms that have a document frequency strictly higher than the given threshold (corpus-specific stop words). The parameter represents a proportion of documents, integer absolute counts. This parameter is ignored if vocabulary is not None.```|
|**min_df : _float in range [0.0, 1.0] or int, default=1_**|
|```When building the vocabulary ignore terms that have a document frequency strictly lower than the given threshold. This value is also called cut-off in the literature. If float, the parameter represents a proportion of documents, integer absolute counts.```|

We can extract our vocabulary along with their summarized TF-IDF scores, inspect the top and bottom 30 terms (same number of terms displayed in LDAvis) and amend the `min_df` and `max_df` values if needed. 

### 1.2. Vocabulary Analysis

````python
def display_top_bottom_terms(tfidf_vectorizer, dtm_tfidf, top_n=30, bottom_n=30):

    # Extract vocabulary and create a dictionary with TF-IDF scores
    vocab = tfidf_vectorizer.get_feature_names_out()
    term_scores = np.array(dtm_tfidf.sum(axis=0)).flatten()
    term_count_dict = dict(zip(vocab, term_scores))
    
    # Highest/Lowest scoring terms
    top_terms = sorted(term_count_dict.items(), key=lambda item: item[1], reverse=True)[:top_n]
    bottom_terms = sorted(term_count_dict.items(), key=lambda item: item[1])[:bottom_n]

    # Create a DataFrame to display side by side
    top_df = pd.DataFrame(top_terms, columns=["Top Term", "TF-IDF Top Score"])
    bottom_df = pd.DataFrame(bottom_terms, columns=["Bottom Term", "TF-IDF Bot Score"])
    combined_df = pd.concat([top_df, bottom_df], axis=1)

    return combined_df
````

````python
display(Markdown("### Top and Bottom 30 Terms by TF-IDF Score"))
print(f"Document-term matrix: {dtm_tfidf.shape[0]} documents x {dtm_tfidf.shape[1]} terms.")
display_side_by_side(display_top_bottom_terms(tfidf_vectorizer, dtm_tfidf)[:12],display_top_bottom_terms(tfidf_vectorizer, dtm_tfidf)[12:])
````

We can see that the first 9 terms with the highest TF-IDF score are generic terms, which are missing from our stop words. We can amend the `max_df` setting of the CountVectorizer, to such a value that these entries are eliminated. Other generic terms can be seen sporadically showing further down such as: _did, say,_ and _really_ which we can manually add to the `stop_words` list. Even in case this does not affect the topic formation greatly, the de-cluttering of the terms presented in the visualisation and will aid the semantic analysis in the limited time allocated.

In terms of low TF-IDF-scoring terms, we can see that `min_df = 10` is a good setting as no terms have an effective score of 0.0, i.e. they are not important enough to form topics. We can see, however, some nonsensical or abbreviated strings which might be the result of type errors or parts of web addresses. In terms of interpretability, similar items might be included in the top 30 terms for a specific topic, which might require the production of a glossary available to the participants to aid them in their analysis.

Through quick-tuning we found the optimal value for `max_df` to be `0.095`. Alongside the revised top - bottom lists we can investigate other words with similar frequencies but different importance which were affected by the change. 


````python
def get_differing_terms(vectorizer1, vectorizer2):
    # Extract vocabularies
    vocab1 = set(vectorizer1.get_feature_names_out())
    vocab2 = set(vectorizer2.get_feature_names_out())

    # Find missing terms
    if len(vocab1) > len(vocab2):
        missing_in_vocab = sorted(vocab1 - vocab2)
    else:
        missing_in_vocab = sorted(vocab2 - vocab1)
    
    # Return the missing terms dataframe
    return pd.DataFrame(missing_in_vocab, columns=["missing term"])
````

````python
re_tfidf_vectorizer  = TfidfVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.095, 
                                min_df = 10)
re_dtm_tfidf  = re_tfidf_vectorizer.fit_transform(newsgroups_df['document'].fillna('')) # replaces na values with empty strings

# Display revised top/bottom terms and the missing terms
display(Markdown("### Re-tuned setting of `max_df = 0.095` and missing terms" ))
print(f"Vocabulary initial term count: {dtm_tfidf.shape[1]}; revised: {re_dtm_tfidf.shape[1]} terms.")
print(f"Document-term matrix: {re_dtm_tfidf.shape[0]} documents x { re_dtm_tfidf.shape[1]} terms.")
display_side_by_side(display_top_bottom_terms(re_tfidf_vectorizer, re_dtm_tfidf),
                     get_differing_terms(tfidf_vectorizer, re_tfidf_vectorizer))
````

To further improve topic interpretability we can also perform some text preprocessing. Lemmatization, i.e. reduce words to their roots, will eliminate multiple forms of the same term; We can also fix the contractions stemming from the use of informal speech. We will use NLTK's [WordNetLemmatizer](https://www.nltk.org/api/nltk.stem.wordnet.html) for the lemmatization and the `contractions` package.

We can then create a new ` document x term ` matrix.

````python
# Map POS tag to wordnet tags
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

def preprocess_text(text, stop_words, exclude_from_lemmatization):
    
    # Fix the contractions (I'm = I am; outta = out of)
    text = contractions.fix(text)
    text = text.lower() # to lowercase
    
    # Remove anything that's not a letter or a white space char
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    tokens = nltk.word_tokenize(text)
    lemmatizer = WordNetLemmatizer()
    processed_tokens = []

    for token in tokens:
        
        # Check if token is in stop words or shorter than 3 characters
        if token in stop_words or len(token) <= 2:
            processed_tokens.append(token)
        else:
            # Lemmatize if not in exclude_from_lemmatization
            if token not in exclude_from_lemmatization:
                token = lemmatizer.lemmatize(token, get_wordnet_pos(token))
            processed_tokens.append(token)
    
    # Join the processed tokens back into a single string
    return ' '.join(processed_tokens)
````

````python
# Create our own stop_words list
initial_word_count = len(re_tfidf_vectorizer.stop_words_)
missing_terms = set(get_differing_terms(tfidf_vectorizer, re_tfidf_vectorizer)['missing term'])

stop_words_to_exclude = {"new", "used", "qur'an", "mars"}
stop_words_to_include = {'edu', 'com', 'problem', 'help', 'use', 'get', 'sure', 'go', 'say', 'read',
                           'do', 'not', 'question', 'point', 'look', 'thing', 'let'} 

custom_stop_words  = (missing_terms - stop_words_to_exclude)
custom_stop_words.update(stop_words_to_include)
custom_stop_words.update(re_tfidf_vectorizer.stop_words_)

exclude_from_lemmatization = {'used', 'windows'} # Car windows or Windows X? 

print(f"Initial stop words count: {initial_word_count}; revised count: {len(custom_stop_words)}.")
````

## 2. Model Fitting

The _SciKitLearn's_ LDA [implementation](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.LatentDirichletAllocation.html) also includes settings for ` doc_topic_prior ` and ` topic_word_prior ` (the prior distributions if known) which we can derive using our pre-existing labels and improve topic coherence. There are a number of other options for fine-tuning the model and improve efficiency alongside interpretability, which could be used if necessary, but would likely fall beyond the scope of this project.

#### Topic Sizes and Position

According to our ground-truth labels topics should be between ~ **3.5%** and **5.5%** of the corpus. Inevitably documents will be miss-labelled, especially when topics are very narrow (e.g. ` comp.harware.ibm-pc ` and ` comp.harware.mac `). Visualisation marks (the topic circles) in the _Intertopic Distance Map_ are positioned closer to each other or further away based on _term-probability overlap._ Specifically the cosine distance between the word vectors.

We can see that topics are decreasing in size, T1 being the largest, followed by T2, T3 and so on. The first three topics are quite larger than what the pre-existing labels would suggest, however we can see that T1 and T3 overlap and suggest that messages in those documents would discuss politics and religion, two topics that have been intertwined throughout human history. T4 however seems to be about sports and is intuitively placed near T1 and T3.

T2 and T5 on the other hand seem to be a generic computing topic and the cryptography topic, with a few noisy terms still present in the smaller topic T5.

On the right-hand side there is a cluster of smaller topics ~ 2% of the corpus in size, which seem to be uninterpretable and are filled with a lot of proper nouns and abbreviations.
````python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 1: Generate sample data
np.random.seed(0)
data = np.random.rand(10, 5)

# Step 2: Perform Agglomerative Clustering
clustering = AgglomerativeClustering(n_clusters=1, linkage='ward', compute_distances=True)
clustering.fit(data)

# The linkage matrix Z (we'll create this using scipy's linkage function)
Z = linkage(data, method='ward')

# Step 3: Extract merge information
# The format of the linkage matrix Z is:
# Z[i] = [idx1, idx2, dist, sample_count]
merge_data = []
for i, (node1, node2, distance, sample_count) in enumerate(Z):
    merge_data.append({
        "node1": int(node1),
        "node2": int(node2),
        "distance": distance
    })

# Print the result
import json
print(json.dumps(merge_data, indent=4))
````

````python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(0)
data = np.random.rand(10, 5)

# Step 2: Perform Agglomerative Clustering and get the linkage matrix
Z = linkage(data, method='ward')

# Step 3: Create a dendrogram to visualize the clustering
plt.figure(figsize=(10, 7))
plt.title("Dendrogram for Agglomerative Clustering")
plt.xlabel("Index")
plt.ylabel("Distance")
dendrogram(Z)
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import linkage, dendrogram

# Example data (replace with your actual data)
np.random.seed(0)
topic_docs_matrix = np.random.rand(20, 5)

# Compute the cosine similarity matrix
cosine_sim_matrix = cosine_similarity(topic_docs_matrix)
df_cosine_sim = pd.DataFrame(cosine_sim_matrix, 
                             columns=[f'Topic {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)],
                             index=[f'Topic {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)])

# Convert the cosine similarity matrix to a distance matrix
distance_matrix = 1 - cosine_sim_matrix

# Perform Agglomerative Clustering using the distance matrix
Z = linkage(distance_matrix, method='ward')

# Create a dendrogram to visualize the clustering
plt.figure(figsize=(10, 7))
plt.title("Dendrogram for Agglomerative Clustering")
plt.xlabel("Topic Index")
plt.ylabel("Distance")
dendrogram(Z, labels=df_cosine_sim.index, leaf_rotation=90)
plt.show()

# Perform MDS to create a 2D mapping
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=0)
mds_coords = mds.fit_transform(distance_matrix)

# Create a DataFrame for the MDS coordinates
df_mds = pd.DataFrame(mds_coords, columns=['Dimension 1', 'Dimension 2'], index=df_cosine_sim.index)

# Plot the 2D mapping
plt.figure(figsize=(10, 7))
plt.scatter(df_mds['Dimension 1'], df_mds['Dimension 2'])

# Annotate points with their topic labels
for label, x, y in zip(df_mds.index, df_mds['Dimension 1'], df_mds['Dimension 2']):
    plt.annotate(label, xy=(x, y), xytext=(5, 5), textcoords='offset points')

plt.title("2D Mapping of Topics Using MDS")
plt.xlabel("X coordinates")
plt.ylabel("Y coordinates")
plt.grid(True)
plt.show()
````
