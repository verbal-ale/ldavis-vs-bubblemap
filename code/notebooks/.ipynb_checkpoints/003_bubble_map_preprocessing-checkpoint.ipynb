# 20 Newsgroups Bubble Map Data Preprocessing

This notebook loads relevant data structures produced by the [application of an LDA model](002_20newsgroups_LDA.ipynb) to the *20 Newsgroups* dataset provided by [scikit-learn](https://scikit-learn.org/stable/datasets/real_world.html#the-20-newsgroups-text-dataset).

It then creates new data structures required as input for the BubbleMap implementation provided by the folks at the *Strategic Futures Laboratory, School of Mathematical and Computer Sciences, Heriot-Watt University, Edinburgh, UK.*

### Notebook Setup & Raw Data Load Up

```python
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import MDS, TSNE

from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from IPython.display import Markdown, display, HTML

# Adjust pandas display settings
pd.set_option('display.max_colwidth', 110)
pd.set_option('display.width', 1250)
pd.set_option('display.colheader_justify', 'left')

# To display two display two dataframes side by side
def display_side_by_side(*args):
    html_str = ''
    for df in args:
        html_str += df.to_html()
    display(HTML('<div style="display: flex; gap: 20px;">' + html_str + '</div>'))
```

```python
document_topic_matrix = np.loadtxt('../../data/document_topic_matrix.csv', delimiter=',')
document_topic_matrix.shape
```

```python
df_topic_term_labels = pd.read_csv('../../data/topic_50_terms_labels_matrix.csv')
df_topic_term_labels.head()
```

```python
df_topic_term_weights = pd.read_csv('../../data/topic_50_terms_weights_matrix.csv')
df_topic_term_weights.head()
```

## 1. Calculate Topic Similarity

We have loaded the document x topic matrix which shows us how much of every of the 20 topics the original document is exhibiting. 

```python
df_doc_topic = pd.DataFrame(document_topic_matrix, columns=[f'Topic {i+1}' for i in range(document_topic_matrix.shape[1])])
df_doc_topic.head(10)
```

We can now transpose the matrix to get the topic definitions by the documents they inhibit. We can then calulate the cosine similarity between those vectors. 

```python
topic_docs_matrix = document_topic_matrix.T
topic_docs_matrix.shape
```

```python
# Calculate cosine similarity between topic vectors
cosine_sim_matrix = cosine_similarity(topic_docs_matrix)
df_cosine_sim = pd.DataFrame(cosine_sim_matrix, 
                             columns=[f'T {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)],
                             index=[f'T {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)])
# Plot a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df_cosine_sim, annot=True, cmap='viridis', vmin=0.05, vmax=0.25, xticklabels=True, yticklabels=True)
plt.title("\nCosine Similarity Matrix\n", fontsize=18, fontweight='bold')
plt.show()
```

## 2. Agglomerative Clustering

Using the cosine similarity matrix we can perform agglomerative clustering and group similar topics together into bigger themes.

```python
# Convert the cosine similarity matrix to a distance matrix
distance_matrix = 1 - cosine_sim_matrix

# Perform Agglomerative Clustering using the distance matrix
linkage_table = linkage(distance_matrix, method='ward')

# Create a dendrogram to visualize the clustering
plt.figure(figsize=(12, 8))
plt.title("\nDendrogram for Agglomerative Clustering\n", fontsize=18, fontweight='bold')
plt.ylabel("Distance")
dendrogram(linkage_table, 
           labels=[f'T {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)], 
           leaf_rotation=90,
           color_threshold=0)
# Set the y-axis limit to start at 0.6
plt.ylim(0.8, plt.ylim()[1])
plt.show()
```

The revised newsgroups were following in one of 5 thematic categories (according to their name definition): 

** **
| theme | number of topics | desc |
|-----|------|------|
| `comp    ` | ` 5 `| ` comp.graphics `, ` comp.hardware.ibm-pc `, ` comp.hardware.mac `, ` comp.windows.misc `, ` comp.windows.x `|
| `misc     `| ` 1 `| ` misc.forsale ` |
| `rec      `| ` 4 `| ` rec.cars `, ` rec.motorcycles `, ` rec.sport.baseball `, ` rec.sport.hockey `|
| `sci     ` | ` 4 `| ` sci.crypt `, ` sci.electronics `, ` sci.med `, ` sci.space ` |
| `talk    ` | ` 6 `| `talk.politics.guns `, ` talk.politics.mideast `, `talk.politics.misc `, ` talk.religion.atheism `, ` talk.religion.christian `, `talk.religion.misc `|

** **

However, after observing the topics using the LDAvis dashboard, we can see that Topic 8 has potentially merged the small topic of ` misc.forsale ` with an automotive/motorcycle one. We can also consider splitting the rather generic theme of "talk" (meaning a newsgroup is dedicated for discussions) into it's subtopics. 

In that case we'll still have 5 themes, however they might be the following: comp, rec, sci, politics, religion. We can examine how many topics each cluster has with either 4 or 5 total themes and also plot dendrograms to visualise the clusters within the hierarchical structure. 

```python
# To print cluster counts for a given number of clusters

def print_cluster_counts(num_clusters, linkage_table):
    
    # Form clusters from the hierarchical clustering
    clusters = fcluster(linkage_table, num_clusters, criterion='maxclust')

    # Create a DataFrame to show the clusters assigned to each topic
    clustered_topics = pd.DataFrame({'Topic': [f'T {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)], 
                                     'C': clusters})

    # Count the number of topics in each cluster
    cluster_counts = clustered_topics['C'].value_counts().sort_index()

    print(f"\nNumber of topics in each cluster for num_clusters = {num_clusters}:")
    print(cluster_counts)
    display_side_by_side(clustered_topics[:10], clustered_topics[10:])

# Print cluster counts for both num_clusters = 4 and 5
print_cluster_counts(4, linkage_table)
print_cluster_counts(5, linkage_table)
```

```python
def show_dendrogram(color_threshold, n_clusters):
    # Plot the dendrogram with color threshold
    plt.figure(figsize=(12, 8))
    plt.title(f"\n {n_clusters} Colored Clusters Dendrogram\n", fontsize=18, fontweight='bold')
    
    dendrogram(linkage_table, labels=[f'T {i}' for i in range(1, topic_docs_matrix.shape[0] + 1)], leaf_rotation=90, color_threshold=color_threshold)
    plt.ylim(0.8, plt.ylim()[1])
    plt.show()

# The color_threshold is set to a value that ensures we get exactly the desired number of clusters
color_threshold = 1.34 # linkage_table[-num_clusters=4, 2] + some error to include T2 and T16 within C4
show_dendrogram(color_threshold, 4)
```

```python
# The color_threshold is set to a value that ensures we get exactly the desired number of clusters
color_threshold = 1.325 # linkage_table[-num_clusters=5, 2] + some error to include T2 and T16 within C4
show_dendrogram(color_threshold, 5)
```

We can see that by introducing a fifth cluster, C2 has now split into ones containing 2 and 4 topics each. The original ` talk. ` cluster, contains 3 and 3 topics of ` religion. ` and ` politics. `, including a ` misc. ` one. 

We'll continue forward using the 5 cluster structure. 

## 3. Calculate 2D mapping

Using the distance matrix and t-SNE embedding we can derive the x/y coordinates for our bubble map.  We'll alseo create the features ` conceptId ` and ` clusterId` to match the naming convention used in the bubble map.

```python
# Remove negative values from distance matrix
non_negative_distance_matrix = np.maximum(0, 1 - cosine_sim_matrix)

# Form flat clusters from the hierarchical clustering
num_clusters = 5
clusters = fcluster(linkage_table, num_clusters, criterion='maxclust')

# Apply t-SNE to project the topics into 2D space
tsne = TSNE(n_components=2, metric='precomputed', perplexity=5, init='random', random_state=0)
tsne_coords = tsne.fit_transform(non_negative_distance_matrix)

# Create a DataFrame for the t-SNE coordinates
df_agglo_clustering = pd.DataFrame(tsne_coords, columns=['x', 'y'])
df_agglo_clustering['conceptId'] = [i for i in range(1, topic_docs_matrix.shape[0] + 1)]
df_agglo_clustering['clusterId'] = clusters

# Plot the 2D t-SNE projection with cluster labels
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_agglo_clustering['x'], df_agglo_clustering['y'], c=df_agglo_clustering['clusterId'], cmap='viridis', s=100)
plt.title("2D t-SNE Projection of Topics")
plt.xlabel("x")
plt.ylabel("y")

# Annotate points with topic labels, with 10 pixels downward offset
for i, topic in enumerate(df_agglo_clustering['conceptId']):
    plt.annotate(topic, (df_agglo_clustering['x'][i], df_agglo_clustering['y'][i]), fontsize=12, ha='right')

# Add a legend for clusters
plt.legend(*scatter.legend_elements(), title="Clusters")

plt.grid(True)
plt.show()
```

## 4. Export JSON data

Finally we need to save our modelling output data as input data for the BubbleMap implementation. 

We need 2 ` .json ` files: one containing **the linkage table** and **topic-term definitions**  and **cluster information** and the other one containing **topic-document distributions** per cluster. 

First we'll export ` newsgroups_sup_map_json.json `.

```python
df_linkage_table = pd.DataFrame(linkage_table[:, :3], columns=["node1", "node2", "distance"])
display(Markdown("#### Topic data and Linkage Table"))
display_side_by_side(df_agglo_clustering, df_linkage_table)
```

```python
conceptData = []

# Iterate over the topics to create JSON objects
for idx, row in df_agglo_clustering.iterrows():
    topic_id = idx
    conceptId = int(row['conceptId'])
    clusterId = int(row['clusterId'])
    x = row['x']
    y = row['y']
    
    # Create the labels array
    labels = []
    for i in range(50):
        label = df_topic_term_labels.iloc[topic_id, i]
        weight = df_topic_term_weights.iloc[topic_id, i]
        labels.append({"label": label, "weight": weight})
    
    # Create the topic JSON object
    topic = {
        "conceptId": conceptId,
        "clusterId": clusterId,
        "hexCoordinates": {
            "x": x,
            "y": y
        },
        "labels": labels
    }
    
    # Add the topic JSON object to the conceptData list
    conceptData.append(topic)

    
# Wrap the conceptData list into a JSON object
newsgroups_sup_map_json = {"conceptsData": conceptData}
newsgroups_sup_map_str = json.dumps(newsgroups_sup_map_json, indent=4)

# Print sample JSON and len of text
display(Markdown("#### JSON file:"))
print(newsgroups_sup_map_str[:500])
print(f"File is {len(newsgroups_sup_map_str)} characters long.")
```

```python
linkage_json = {"linkageTable": df_linkage_table.to_dict(orient="records")}
newsgroups_sup_map_json.update(linkage_json)
newsgroups_sup_map_str = json.dumps(newsgroups_sup_map_json, indent=4)
print(f"File is now {len(newsgroups_sup_map_str)} characters long.")
```

```python
# Save the updated JSON string to a file
with open('../../data/newsgroups_sup_map.json', 'w') as json_file:
    json_file.write(newsgroups_sup_map_str)
```

Finally we need to export the accumulative topic sizes in terms of their distributions across the documents in a file called ` newsgroups_sup_distribution.json `.

```python
document_topic_matrix.shape
```

```python
column_sums = np.sum(document_topic_matrix, axis=0)
column_sums
```

```python
df_numberOfDocs_topics = pd.DataFrame(column_sums, columns=["numberOfDocuments"])
df_numberOfDocs_topics = df_numberOfDocs_topics.sort_values(
    by="numberOfDocuments", ascending=False).reset_index(drop=True)
df_numberOfDocs_topics
```

```python
# Convert the DataFrame to a JSON structure
json_data = []
for idx, row in df_numberOfDocs_topics.iterrows():
    json_obj = {
        "conceptId": idx + 1,
        "topicData": {
            "numberOfDocuments": row["numberOfDocuments"],
            "label": "Total Documents"
        }
    }
    json_data.append(json_obj)

# Convert to JSON string
json_str = json.dumps(json_data, indent=4)
print(json_str[:310], "\n...")
```

```python
# Save the JSON string to a file
with open('../../data/newsgroups_sup_distrib.json', 'w') as f:
    f.write(json_str)
```
